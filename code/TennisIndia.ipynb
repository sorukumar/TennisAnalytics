```python


# Define the order of rounds
round_order = {'R128': 1, 'R64': 2, 'R32': 3, 'R16': 4, 'QF': 5, 'SF': 6, 'F': 7, 'W': 8}
grand_slams = ['Australian Open', 'French Open', 'Wimbledon', 'US Open']

def get_best_performance(group):
    group['round_order'] = group['round'].map(round_order)
    group = group.dropna(subset=['round_order'])  # Ensure this row to remove any NaNs before using idxmax
    if group.empty:
        return pd.DataFrame({'best_performance': [None], 'tourney_name': [None]})
    best_row = group.loc[group['round_order'].idxmax()]  # Now it should not encounter NaN
    return pd.DataFrame({'best_performance': [best_row['round']], 'tourney_name': [best_row['tourney_name']]})


def get_player_details(atp_data, country_code):
    # Filter matches involving players from the specified country
    country_players = atp_data[atp_data['winner_ioc'].eq(country_code) | atp_data['loser_ioc'].eq(country_code)]

    # Collect player details for winners and losers
    winners = country_players.loc[country_players['winner_ioc'] == country_code, ['winner_name', 'winner_id', 'year', 'round', 'tourney_name']].rename(columns={'winner_name': 'name', 'winner_id': 'id'})
    losers = country_players.loc[country_players['loser_ioc'] == country_code, ['loser_name', 'loser_id', 'year', 'round', 'tourney_name']].rename(columns={'loser_name': 'name', 'loser_id': 'id'})
    
    all_players = pd.concat([winners, losers])
    # Filter out Grand Slam matches
    grand_slam_matches = all_players[all_players['tourney_name'].isin(grand_slams)]
    
    # Calculate the best performance each year and the best lifetime performance
    best_performance_each_year = all_players.groupby(['id', 'year']).apply(lambda x: get_best_performance(x), include_groups=False).reset_index()
    best_performance_lifetime = all_players.groupby(['id']).apply(lambda x: get_best_performance(x), include_groups=False).reset_index()

    # Calculate the best Grand Slam performance for each player
    best_grand_slam_performance = grand_slam_matches.groupby(['id']).apply(lambda x: get_best_performance(x), include_groups=False).reset_index()
    best_grand_slam_performance.rename(columns={'best_performance': 'best_grand_slam_performance', 'tourney_name': 'best_grand_slam_tourney'}, inplace=True)


    # Aggregate names and years based on player ID
    player_details = all_players.groupby('id').agg({
        'name': 'first',  
        'year': lambda x: sorted(set(x))
    }).reset_index()

    # Merge the best lifetime performance with player details
    player_details = player_details.merge(best_performance_lifetime, on='id', how='left')
    player_details.rename(columns={'best_performance': 'best_performance_lifetime', 'tourney_name': 'best_tourney_lifetime'}, inplace=True)

    # Format best_performance_lifetime to include round information
    player_details['best_performance_lifetime'] = player_details.apply(
        lambda row: f"{row['best_performance_lifetime']} at {row['best_tourney_lifetime']}" if row['best_performance_lifetime'] and row['best_tourney_lifetime'] else None, axis=1
    )

    # Merge the best Grand Slam performance details back into the player details
    player_details = player_details.merge(best_grand_slam_performance[['id', 'best_grand_slam_performance', 'best_grand_slam_tourney']], on='id', how='left')
    # Optionally format the Grand Slam performance for readability
    player_details['best_grand_slam_performance'] = player_details.apply(
        lambda row: f"{row['best_grand_slam_performance']} at {row['best_grand_slam_tourney']}" if pd.notna(row['best_grand_slam_performance']) else 'No Grand Slam Matches', axis=1
        )

    # Calculate the number of matches played
    match_counts = country_players.melt(id_vars=['year'], value_vars=['winner_id', 'loser_id'], var_name='role', value_name='id').groupby(['id', 'year']).size().reset_index(name='matches_played')

    # Expand the years into separate rows for detailed merging
    player_details_expanded = player_details.explode('year')

    # Merge the best performance and match counts with expanded player details
    final_details = player_details_expanded.merge(best_performance_each_year, on=['id', 'year'], how='left').merge(match_counts, on=['id', 'year'], how='left')

    # Combine best performance and tourney name by year into a single string
    final_details['best_performance_each_year'] = final_details.apply(
        lambda row: f"{row['year']}: {row['best_performance']} at {row['tourney_name']}" if row['best_performance'] and row['tourney_name'] else None, axis=1
    )

    # Now 'best_performance_lifetime' is formatted with rounds and tourney names, and included in the final aggregation


    # Group back by player ID and aggregate the results
    final_player_details = final_details.groupby('id').agg({
        'name': 'first',
        'year': lambda x: sorted(set(x)),
        'matches_played': 'sum',
        'best_performance_each_year': lambda x: ', '.join(filter(None, x)),
        'best_performance_lifetime': 'first',  
        'best_tourney_lifetime': 'first',
        'best_grand_slam_performance': 'first',  # Add this line
        'best_grand_slam_tourney': 'first'       # And this line
    }).reset_index()

    return final_player_details[['id', 'name', 'year', 'matches_played', 'best_performance_each_year', 'best_performance_lifetime', 'best_tourney_lifetime', 'best_grand_slam_performance', 'best_grand_slam_tourney']]

```

```python

def count_total_unique_players(atp_data, country_code):
    # Filter players from the specified country who either won or lost
    winners = atp_data[atp_data['winner_ioc'] == country_code]
    losers = atp_data[atp_data['loser_ioc'] == country_code]

    # Create sets of unique player names from winners and losers
    country_unique_players = set(winners['winner_name']).union(set(losers['loser_name']))
    global_unique_players = set(atp_data['winner_name']).union(set(atp_data['loser_name']))

    # Return a DataFrame with counts
    results_df = pd.DataFrame({
        'Total Unique Players': [len(country_unique_players), len(global_unique_players)],
        'Country': [country_code, 'Global']
    })
    return results_df



def count_players_per_year(atp_data, country_code):
    # Filter players from the specified country who either won or lost
    winners = atp_data[atp_data['winner_ioc'] == country_code]
    losers = atp_data[atp_data['loser_ioc'] == country_code]

    # Combine filtered data and count unique players per year for the country
    combined_country_players = pd.concat([winners, losers])
    country_unique = combined_country_players.groupby('year').agg({'winner_name': pd.Series.nunique, 'loser_name': pd.Series.nunique}).sum(axis=1)
    
    # Count unique players per year globally
    global_unique = atp_data.groupby('year').agg({'winner_name': pd.Series.nunique, 'loser_name': pd.Series.nunique}).sum(axis=1)

    # Combine results into one DataFrame
    combined = pd.DataFrame({
        'Total Players from ' + country_code: country_unique,
        'Total Players Globally': global_unique
    }).fillna(0)

    return combined



def win_loss_record(atp_data, country_code):
    # Consider matches where either winner or loser is from the specified country
    matches = atp_data[(atp_data['winner_ioc'] == country_code) | (atp_data['loser_ioc'] == country_code)]

    # Wins and losses per year for the country
    wins = matches[matches['winner_ioc'] == country_code].groupby('year').size()
    losses = matches[matches['loser_ioc'] == country_code].groupby('year').size()
    total_matches = atp_data.groupby('year').size()

    # Preparing the DataFrame just before returning
    result_df = pd.DataFrame({
        'Wins from ' + country_code: wins,
        'Losses from ' + country_code: losses,
        'Total Matches': total_matches
    }).fillna(0)

    return result_df



def tournament_participation_by_level(atp_data, country_code):
    # Filter matches involving players from the specified country
    tournaments = atp_data[(atp_data['winner_ioc'] == country_code) | (atp_data['loser_ioc'] == country_code)]

    # Count tournaments by level per year for the country and globally
    country_participation_by_level = tournaments.groupby(['year', 'tourney_level']).size().unstack(fill_value=0)
    global_participation_by_level = atp_data.groupby(['year', 'tourney_level']).size().unstack(fill_value=0)

    return country_participation_by_level, global_participation_by_level

```
