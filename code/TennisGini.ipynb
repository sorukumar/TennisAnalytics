```python
# Define the order of rounds
round_order = {'R128': 1, 'R64': 2, 'R32': 3, 'R16': 4, 'QF': 5, 'SF': 6, 'F': 7, 'W': 8}

def get_player_details(atp_data):
    # Collect player details for winners and losers
    winners = atp_data[['winner_name', 'winner_id', 'year', 'round', 'winner_ioc']].rename(columns={'winner_name': 'name', 'winner_id': 'id', 'winner_ioc': 'country'})
    losers = atp_data[['loser_name', 'loser_id', 'year', 'round', 'loser_ioc']].rename(columns={'loser_name': 'name', 'loser_id': 'id', 'loser_ioc': 'country'})
    
    all_players = pd.concat([winners, losers])
    
    # Aggregate names and years based on player ID
    player_details = all_players.groupby('id').agg({
        'name': 'first',  
        'year': lambda x: sorted(set(x)),
        'round': list,
        'country': 'first'
    }).reset_index()

    # Calculate the number of matches played
    match_counts = all_players.groupby('id').size().reset_index(name='matches_played')

    # Calculate the number of trophies
    trophies = winners[winners['round'] == 'F'].groupby('id').size().reset_index(name='# of trophies')
    
    # Calculate the number of finals, semi-finals, quarter-finals, and round of 16
    performance_counts = all_players.groupby(['id', 'round']).size().unstack(fill_value=0).reset_index()
    
    # Combine the details
    final_player_details = player_details.merge(match_counts, on='id', how='left').merge(performance_counts, on='id', how='left').merge(trophies, on='id', how='left')
    final_player_details.fillna(0, inplace=True)
    
    # Calculate additional metrics
    final_player_details['years_active'] = final_player_details['year'].apply(len)
    final_player_details['min_year'] = final_player_details['year'].apply(min)
    final_player_details['max_year'] = final_player_details['year'].apply(max)

    final_player_details['# of Final'] = final_player_details['F'] if 'F' in final_player_details.columns else 0
    final_player_details['# of Semi final'] = final_player_details['SF'] if 'SF' in final_player_details.columns else 0
    final_player_details['# of quarterfinal'] = final_player_details['QF'] if 'QF' in final_player_details.columns else 0
    final_player_details['# of round 16'] = final_player_details['R16'] if 'R16' in final_player_details.columns else 0

    return final_player_details[['name', 'country', 'matches_played', '# of trophies', '# of Final', '# of Semi final', '# of quarterfinal', '# of round 16', 'years_active', 'min_year', 'max_year']]

```

```python
# Example usage:
players_info = get_player_details(atp_data)
```

```python
# Filter data to include only players who started their careers in 1969 or later
filtered_data = players_info[players_info['min_year'] >= 1969].copy()
filtered_data['years_played'] = filtered_data['max_year'] - filtered_data['min_year'] + 1

# Calculate the cumulative percentage of players lasting different durations
duration_counts = filtered_data['years_played'].value_counts().sort_index()
cumulative_counts = duration_counts.cumsum()
total_players = filtered_data.shape[0]
cumulative_percentage = cumulative_counts / total_players * 100

# Prepare data for visualization
career_duration_data = pd.DataFrame({
    'Years Played': duration_counts.index,
    'Number of Players': duration_counts.values,
    'Cumulative Percentage': cumulative_percentage.values
})

# Calculate cumulative percentage for trophies
trophy_counts = filtered_data['# of trophies'].value_counts().sort_index()
cumulative_percentage_trophies = trophy_counts.cumsum() / trophy_counts.sum() * 100

# Prepare data for the second visualization
trophy_data = pd.DataFrame({
    'Trophies Won': trophy_counts.index,
    'Number of Players': trophy_counts.values,
    'Cumulative Percentage': cumulative_percentage_trophies.values
})
```

```python
# Sort the players by 'number of trophies' in descending order
sorted_data = filtered_data.sort_values(by='# of trophies', ascending=False)
sorted_data['cumulative_trophies'] = sorted_data['# of trophies'].cumsum()
sorted_data['cumulative_percentage_trophies'] = sorted_data['cumulative_trophies'] / sorted_data['# of trophies'].sum() * 100

# Calculate the cumulative number of players (since each row represents one player)
sorted_data['cumulative_players'] = np.arange(1, len(sorted_data) + 1)
# Calculate the cumulative percentage of players
sorted_data['cumulative_percentage_players'] = sorted_data['cumulative_players'] / len(sorted_data) * 100
final_sorted_data = sorted_data[['name', 'country', 'min_year', 'cumulative_trophies', 
                                 'cumulative_percentage_trophies', 'cumulative_players', 
                                 'cumulative_percentage_players']]

```

```python

# Define colors from Category20 palette
bar_color = to_rgba('C0', alpha=0.75)  # Blue shade
line_color = 'C3'  # Red shade
highlight_color = 'C2'  # Green shade
annotation_color = 'C7'  # Grey shade

# Create the bar chart (number of players) and line graph (cumulative percentage)
fig1, ax1 = plt.subplots(figsize=(14, 8))

# Bar chart
bars = ax1.bar(career_duration_data['Years Played'], career_duration_data['Number of Players'], color=bar_color, edgecolor=bar_color)
ax1.set_xlabel('Number of Years Played')
ax1.set_ylabel('Number of Players')
ax1.set_title('Distribution of ATP Career Longevity Durations (Starting from 1969)',  pad=20)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Line chart for cumulative percentage
ax2 = ax1.twinx()
ax2.plot(career_duration_data['Years Played'], career_duration_data['Cumulative Percentage'], color=line_color, marker='o', linestyle='-', linewidth=2)
ax2.set_ylabel('Cumulative Percentage')
ax2.set_ylim(0, 100)
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x)}%'))

# Adding annotations for key points
highlight_points = [1, 2, 5, 15]
for point in highlight_points:
    cumulative_percent = career_duration_data.loc[career_duration_data['Years Played'] == point, 'Cumulative Percentage'].values[0]
    ax2.annotate(f'{cumulative_percent:.0f}%', xy=(point, cumulative_percent), xytext=(point, cumulative_percent + 5),
                 arrowprops=dict(facecolor=annotation_color, shrink=0.05), ha='center', color=annotation_color)

# Highlighting the percentage thresholds
thresholds = [80, 90, 95]
for threshold in thresholds:
    years_at_threshold = career_duration_data[career_duration_data['Cumulative Percentage'] >= threshold].iloc[0]['Years Played']
    ax2.axhline(y=threshold, color=highlight_color, linestyle='--', linewidth=1)
    ax2.axvline(x=years_at_threshold, color=highlight_color, linestyle='--', linewidth=1)
    ax2.annotate(f'{threshold}%', xy=(years_at_threshold, threshold), xytext=(years_at_threshold + 1, threshold - 5),
                 arrowprops=dict(facecolor=highlight_color, shrink=0.05), ha='left', color=highlight_color)

plt.show()
```

```python

fig3, ax = plt.subplots(figsize=(14, 8))

# Plot cumulative percentage of players vs. cumulative percentage of trophies
ax.plot(final_sorted_data['cumulative_percentage_players'], final_sorted_data['cumulative_percentage_trophies'], marker='o', linestyle='-')

# Setting the labels
ax.set_xlabel('Cumulative Percentage of Players (%)')
ax.set_ylabel('Cumulative Percentage of Trophies (%)')

# Adding a grid for better readability
ax.grid(True)

# Calculate and annotate specific percentages
percentages = [1, 5, 9]
colors = ['red', 'green', 'blue']
labels = ['1% of players hold', '5% of players hold', '9% of players hold']

for percent, color, label in zip(percentages, colors, labels):
    # Find the nearest point to the specific percentage of players
    nearest_value = final_sorted_data.iloc[(final_sorted_data['cumulative_percentage_players']-percent).abs().argsort()[:1]]
    x_value = nearest_value['cumulative_percentage_players'].values[0]
    y_value = nearest_value['cumulative_percentage_trophies'].values[0]

    # Determine optimal text positioning to avoid overlapping
    # Adjust text position based on the percentage to avoid going outside the chart
    offset_x = 10 if x_value < 50 else -120  # Adjust x offset based on position to avoid going outside the chart
    offset_y = 10 if y_value < 50 else -10  # Adjust y offset to avoid covering the line
    
    # Annotate on the plot with adjusted positions
    ax.annotate(f'{label} {y_value:.0f}% of trophies', xy=(x_value, y_value), xytext=(offset_x, offset_y),
                textcoords='offset points', arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color=color), color=color)

# Adding a title
ax.set_title('Lorenz Curve for Trophy Distribution Among Players (starting from 1969). Total players 6K, Total Trophies 3.6K')

# Show plot
plt.show()

```
